Extract more information for comparison from the table by creating month and year columns

```python
rows = []
for i in range(len(netflixDf)):
    if netflixDf['date_added'].iloc[i] == '':
        rows.append(i)
netflixDf.loc[rows,:]
```

```python
month_added = []
year_added = []
for i in range(len(netflixDf)):
    if i in rows:
        month_added.append(0)
        year_added.append(0)
    else:
        date = netflixDf['date_added'].iloc[i].split(' ')
        month_added.append(date[0])
        year_added.append(int(date[2]))
        
for i , month in enumerate(month_added):
    if month != 0:
        datetime_obj = datetime.strptime(month, '%B')
        month_number = datetime_obj.month
        month_added[i] = month_number

netflixDf.insert(7,'month_added',month_added, allow_duplicates=True)
netflixDf.insert(8,'year_added',year_added, allow_duplicates=True)
netflixDf.head(7)
```

getDuration function to specify TV Show , Movies length 

```python
def getDuration(data):
    count = 0
    durations = []
    for value in data:
	# filling in missing values
        if type(value) is float:
            durations.append(0)
        else:
            values = value.split(" ")
            durations.append(int(values[0]))
    return durations
```

Function to find unique values in a column

```python
def uniqueItem(data):
    unique_value = set()
    for value in data:
        if type(value) is float:
            unique_value.add(None)
        else:
            value = value.split(', ')
            for i in range(len(value)):
                unique_value.add(value[i])
    return list(unique_value)
    
    
```
